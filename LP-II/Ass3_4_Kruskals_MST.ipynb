```python
from operator import itemgetter

class MinHeap:
    def __init__(self):
        self.heap = []
    
    def heapify(self,idx):
        left = (2*idx + 1)
        right = (2*idx + 2)
        small = idx
        if left < self.size() and self.heap[left][0] < self.heap[small][0]:
            small = left
        if right < self.size() and self.heap[right][0] < self.heap[small][0]:
            small = right
        
        
        if small != idx:
            temp = self.heap[idx]
            self.heap[idx] = self.heap[small]
            self.heap[small] = temp
            self.heapify(small)
    
    def push(self,value):
        insertedIdx = self.size()
        self.heap.append(value)
        parentIdx = int((insertedIdx-1)/2)
        
        while parentIdx >= 0 and self.heap[parentIdx][0] > self.heap[insertedIdx][0]:
            temp = self.heap[parentIdx]
            self.heap[parentIdx] = self.heap[insertedIdx]
            self.heap[insertedIdx] = temp
            insertedIdx = parentIdx
            parentIdx = int((insertedIdx-1)/2)
    
    def pop(self):
        heapSize = self.size()
        if  heapSize <= 0 :
            return None
        top = self.heap[0]
        self.heap[0] = self.heap[heapSize-1]
        self.heap.pop()
        self.heapify(0)
        return top
    
    
    def size(self):
        return len(self.heap)
    
class DisJointSet:
    
    def __init__(self,v):
        self.parent = [0]*v
        self.size = [0]*v
        
        for i in range(v):
            self.parent[i] = i
            self.size[i] = 1
    
    def findParent(self,node):
        if self.parent[node] == node:
            return node
        
        p = self.findParent(self.parent[node])
        self.parent[node] = p
        return p 
    
    def unionBySize(self,u,v):
        ultp_u = self.findParent(u)
        ultp_v = self.findParent(v)
        
        if(ultp_u == ultp_v):
            return
        
        if self.size[ultp_u] > self.size[ultp_v]:
            self.parent[ultp_v] = ultp_u
            self.size[ultp_u] += self.size[ultp_v]
        else:
            self.parent[ultp_u] = ultp_v
            self.size[ultp_v] += self.size[ultp_u]
            
            
if __name__ == '__main__':
    v = int(input("Enter the number of vertices : "))
    
    # create an empty graph
    graph = [[0 for i in range(v)] for j in range(v)]
    
    e = int(input("Enter the number of edges : "))
    # take edges with weight and add in graph
    dist = [1e9]*v
    minHeap = MinHeap()
    st = DisJointSet(v)
    # edges = []
    for i in range(e):
        input_line = input("Enter edge %d (s, d, w):"%(i+1))
        input_line = input_line.split();
        input_line = [int(ele) for ele in input_line]
        s = input_line[0]
        d = input_line[1]
        w = input_line[2]
        minHeap.push([w,[s,d]])
    #  edges.append([w,[s,d]])
    
    #  edges =  sorted(edges, key=itemgetter(0))
    count = 0
    cost = 0
    while minHeap.size() > 0 and count < v :
        edge = minHeap.pop()
        w = edge[0]
        s = edge[1][0]
        d = edge[1][1]
        
#         print(f"src: {s}  dist: {d} wt: {w}")
        if st.findParent(s) != st.findParent(d):
            st.unionBySize(s,d)
            cost += w
            count+=1
    
    print('Cost : ',cost)
```
