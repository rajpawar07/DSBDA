```python
class MinHeap:
    def __init__(self):
        self.heap = []
        
    def size(self):
        return len(self.heap)
        
    def heapify(self,idx):
        left = idx*2+1
        right = idx*2+2
        small = idx
        if left < self.size() and self.heap[left][0] < self.heap[idx][0]:
            small = left
            
        if right < self.size() and self.heap[right][0] < self.heap[idx][0]:
            small = right
        
        if small != idx:
            temp = self.heap[small]
            self.heap[small] = self.heap[idx]
            self.heap[idx] = temp
        
            self.heapify(small)
        
        
    def push(self,value):
        newElementIdx = len(self.heap);
        self.heap.append(value)
        
        parentIdx = int((newElementIdx-1)/2)
        # perform heap up operation
        while parentIdx >= 0 and self.heap[parentIdx][0] > self.heap[newElementIdx][0]:
            temp = self.heap[parentIdx]
            self.heap[parentIdx] = self.heap[newElementIdx]
            self.heap[newElementIdx] = temp
            newElementIdx = parentIdx
            parentIdx = int((newElementIdx-1)/2)
            
    def pop(self):
        if self.size() <= 0:
            return None
        
        top = self.heap[0]
        self.heap[0] = self.heap[self.size()-1]
        self.heap.pop()
        self.heapify(0)
        return top

if __name__ == '__main__':
    v = int(input("Enter the number of vertices : "))
    
    # create an empty graph
    graph = [[0 for i in range(v)] for j in range(v)]
    print(graph)
    e = int(input("Enter the number of edges : "))
    # take edges with weight and add in graph
    for i in range(e):
        input_line = input("Enter edge %d (s, d, w):"%(i+1))
        input_line = input_line.split();
        input_line = [int(ele) for ele in input_line]
        s = input_line[0]
        d = input_line[1]
        w = input_line[2]
        graph[s][d] = w
        graph[d][s] = w
    
    dist = [1e9]*v
    mst = [0]*v
    minHeap = MinHeap()
    minHeap.push([0,[0]])
    parent = [-1]*v
    
    dist[0] = 0
    mst[0] = 1
    
    # There are v vertices in MST and souce is already included in MST
    while minHeap.size() > 0 :
        node = minHeap.pop();
        u = node[1][0]
        mst[u] = 1;
        
        for adj in range(v):
            if graph[u][adj] != 0:
                wt = graph[u][adj]
                if mst[adj] == 0 and wt < dist[adj]:
                    parent[adj] = u
                    dist[adj] = wt
                    minHeap.push([dist[adj],[adj]])
        
    
    cost = 0;
    for wt in dist:
        cost += wt
    
    print(dist)
    print(cost)
```
